```python
from unstructured_client import UnstructuredClient
from unstructured_client.models import shared
from unstructured_client.models.errors import SDKError

from unstructured.partition.html import partition_html
from unstructured.partition.pdf import partition_pdf

from unstructured.staging.base import dict_to_elements
```

```python
s=UnstructuredClient(api_key_auth='')
```

```python
filename = "/content/POSTIFFERI Audition.pdf"
pdf_elements = partition_pdf(filename=filename, strategy="fast")
```

```python
filename = "/content/POSTIFFERI Audition.pdf"

with open(filename, "rb") as f:
    files=shared.Files(
        content=f.read(),
        file_name=filename,
    )
```

```python
req = shared.PartitionParameters(
    files=files,
    strategy="hi_res",
    hi_res_model_name="detectron2_onnx",
)

try:
    resp = s.general.partition(req)
    dld_elements = dict_to_elements(resp.elements)
except SDKError as e:
    print(e)
```

```python
for element in dld_elements:
    print(f"Page Number : {element.metadata.page_number}: {element.text}")
```

```python
from langchain_community.document_loaders import UnstructuredAPIFileLoader
```

```python
loader = UnstructuredAPIFileLoader(
    file_path=filename,
    api_key="",
    mode='paged'
)
```
